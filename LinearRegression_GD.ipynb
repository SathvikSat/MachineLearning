```python
def train(X, Y, lr = 0.05, epochs = 100):
    W = np.random.rand(1)
    b = np.random.rand(1)
    for e in range(epochs):
        h = np.dot(X , W) + b
        err = (Y - h)
        loss = np.mean(a = err**2)
        W[0] = W[0] - lr * (np.mean((X[:,0]) * err.T))
        
        b = b - lr * (np.mean(err))
        
        if e % 50 == 0:
            print("epochs: ",e, "loss: ",loss)
    return W, b        
            
```

```python
W, b = train(X_train, y_train, lr = 0.01, epochs= 200)
```
